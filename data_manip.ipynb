```python
import pandas as pd

# 1. Load the CSV
df = pd.read_csv("/Users/noah/NMR_filtering-1/cs_stat_aa_filt.csv")

df["lower_5sd"] = df["avg"] - 5 * df["std"]
df["upper_5sd"] = df["avg"] + 5 * df["std"]

df["range_5sd"] = "[" + df["lower_5sd"].round(3).astype(str) + ", " + df["upper_5sd"].round(3).astype(str) + "]"


df = df[df["atom_id"].str.startswith("C")]


df["atom_id"].unique().tolist()

```

```python
df.sort_values(by = "atom_id", inplace = True)

print(df["comp_id"].nunique())
print(df["atom_id"].nunique())
df.head()
```

```python
CA_DF = df[df["atom_id"] == "ILE"].copy()
CA_DF.sort_values(by = "upper_5sd", inplace = True, ascending = True)

print(CA_DF["comp_id"].nunique())
CA_DF.head(50)

```
